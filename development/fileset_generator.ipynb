```python
import json
```

```python
def getraw(jsonfilename):
    '''
    Load the raw dictionary
    '''
    with open(jsonfilename) as f :
        full_fileset = json.load(f)
    return full_fileset
```

```python
zh = 'p8_ee_ZH_ecm240'
zz = 'p8_ee_ZZ_ecm240'
ww = 'p8_ee_WW_ecm240'
path = '/eos/experiment/fcc/ee/generation/DelphesEvents/spring2021/IDEA/'
filename = 'fileset.json'
```

```python

with open('./'+zh+'.txt') as f_zh :
    lines_ZH = f_zh.readlines()
    lines_ZH = [line.strip() for line in lines_ZH]
with open('./'+zz+'.txt') as f_zz :
    lines_ZZ = f_zz.readlines()
    lines_ZZ = [line.strip() for line in lines_ZZ]
with open('./'+ww+'.txt') as f_ww :
    lines_WW = f_ww.readlines()
    lines_WW = [line.strip() for line in lines_WW]

print(lines_ZH[0])
print(lines_ZZ[0])
print(lines_WW[0])
```

```python
dataset = {zh:lines_ZH,zz:lines_ZZ,ww:lines_WW}
dict = {}
for key in dataset.keys():
    file_dict = {path+key+'/'+file:'events' for file in dataset[key]}
    dict[key] = {'files': file_dict}
dict
```

```python
with open(filename, 'w') as f :
    json.dump(dict, f)
```
